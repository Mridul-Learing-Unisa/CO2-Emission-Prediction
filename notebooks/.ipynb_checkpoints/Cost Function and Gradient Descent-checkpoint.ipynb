```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys

sys.path.append("../Scripts")

from data_processing import get_data
from calculate_w_and_b import get_w_b
from calculate_costFunction import compute_cost
from Gradient_Descent import *
```

```python
df = pd.read_csv("../DataSets/CO2 Emissions_Canada.csv")
```

```python
X, y = get_data("../DataSets/CO2 Emissions_Canada.csv")

x_train = X['Fuel Consumption'].to_numpy()
y_train = y.to_numpy()

w,b = get_w_b("../DataSets/CO2 Emissions_Canada.csv")


mean_x = np.mean(x_train)
std_x = np.std(x_train)
x_norm = (x_train - mean_x) / std_x

print(x_norm)
```

```python
# initialize parameters
w_init = 0
b_init = 0

# some gradient descent settings
iterations = 10000
tmp_alpha = 1e-4 

# run gradient descent
w_final, b_final, J_hist, p_hist = gradient_descent(x_norm ,y_train, w_init, b_init, tmp_alpha, iterations, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```

```python
plt.plot(J_hist[:100])
plt.title("Cost vs. iteration(start)");
plt.ylabel('Cost')           
plt.xlabel('iteration step')
plt.show()
```
