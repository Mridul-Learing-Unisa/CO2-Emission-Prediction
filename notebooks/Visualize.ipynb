```python
w, b = get_w_b("../DataSets/CO2 Emissions_Canada.csv")
```

```python
X, y = get_data("../DataSets/CO2 Emissions_Canada.csv")

x_train = X['Fuel Consumption'].to_numpy()
y_train = y.to_numpy()
```

```python
def calculate_model_output(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b
    return f_wb
        
```

```python
model_prediction = compute_model_output(x_train,w,b)
```

```python
plt.scatter(x_train, y_train, marker = 'x', c='r')
plt.title("CO2 Emission vs Fuel Consumptiom")
plt.xlabel("Fuel Consumptio")
plt.ylabel("CO2 Emission")
plt.plot(x_train, model_prediction,c='b')
```
